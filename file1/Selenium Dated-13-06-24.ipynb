```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
import time
import re
import requests
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, ElementClickInterceptedException
```

```python
options = Options()
options.add_experimental_option("detach", True) # To keep the browser open 

driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()), options=options) # Initializing the web driver according to the chrome version 

driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
time.sleep(3)

ranks = []
names = []
artists = []
upload_dates = []
views = []

# Name

pattern = r'\[.*\]'

count = 1

try :
    name = driver.find_elements(By.XPATH, "//table[@class='sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter']/tbody/tr/td[1]")
    for i in name:
        names.append(re.sub(pattern, "", i.text))
        ranks.append(count)
        count = count + 1
        
except NoSuchElementFound:
    print("Try to change the path.")
    
# Artists

try :
    artist = driver.find_elements(By.XPATH, "//table[@class='sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter']/tbody/tr/td[2]")
    for i in artist:
        artists.append(i.text)
        
except NoSuchElementFound:
    print("Try to change the path.")
        
    
# Views

try :
    view = driver.find_elements(By.XPATH, "//table[@class='sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter']/tbody/tr/td[3]")
    for i in view:
        views.append(i.text)
        
except NoSuchElementFound:
    print("Try to change the path.")
    
# Upload_Dates

try :
    upload_date = driver.find_elements(By.XPATH, "//table[@class='sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter']/tbody/tr/td[4]")
    for i in upload_date:
        upload_dates.append(i.text)
        
except NoSuchElementFound:
    print("Try to change the path.")
    
```

```python
options = Options()
options.add_experimental_option("detach", True) # To keep the browser open 

driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()), options=options) # Initializing the web driver according to the chrome version 

driver.get('https://archive.ics.uci.edu/')
time.sleep(3)
```

```python
dataset_button = driver.find_element(By.XPATH, "/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]")
dataset_button.click()
time.sleep(3)
```

```python
links = []

time.sleep(3)
while True:
    try :
        a = driver.find_elements(By.XPATH, "//a[@class='link-hover link text-xl font-semibold']")
        for i in a:
            links.append(i.get_attribute('href'))
            
        try :
            next_button = driver.find_element(By.XPATH, "/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]")
            next_button.click()
            time.sleep(2)
            
        except ElementClickInterceptedException:
            print("All the data has been scraped")
            break
            
    except NoSuchElementException:
        print("Change the path")
```

```python
len(links)
```

```python
dataset_names = []
data_types = []
tasks = []
attribute_types = []
instances = []
attributes = []
years = []

try :
    for i in links:
        driver.get(i)
        
        # Dataset Names

        try :
            a = driver.find_element(By.XPATH, "//h1[@class='text-3xl font-semibold text-primary-content']")
            dataset_names.append(a.text)
        
        except NoSuchElementException:
            dataset_names.append("-")
    
        # Data Types

        try :
            a = driver.find_element(By.XPATH, "//div[@class='grid grid-cols-8 gap-4 md:grid-cols-12']/div[4]/p")
            data_types.append(a.text)
        
        except NoSuchElementException:
            data_types.append("-")
            
        # Task
        
        try :
            a = driver.find_element(By.XPATH, "//div[@class='grid grid-cols-8 gap-4 md:grid-cols-12']/div[3]/p")
            tasks.append(a.text)
        
        except NoSuchElementException:
            tasks.append("-")
            
        # Attribute Type
        
        try :
            a = driver.find_element(By.XPATH, "//div[@class='grid grid-cols-8 gap-4 md:grid-cols-12']/div[1]/p")
            attribute_types.append(a.text)
        
        except NoSuchElementException:
            attribute_types.append("-")   
            
        # No of Instances
        
        try :
            a = driver.find_element(By.XPATH, "//div[@class='grid grid-cols-8 gap-4 md:grid-cols-12']/div[5]/p")
            instances.append(a.text)
        
        except NoSuchElementException:
            instances.append("-")
            
        # No of Atrributes
        
        try :
            a = driver.find_element(By.XPATH, "//div[@class='grid grid-cols-8 gap-4 md:grid-cols-12']/div[6]/p")
            attributes.append(a.text)
        
        except NoSuchElementException:
            attributes.append("-")
            
        # Year
        
        try :
            a = driver.find_element(By.XPATH, "//h2[@class='text-sm text-primary-content']")
            years.append(a.text)
        
        except NoSuchElementException:
            years.append("-")

except NoSuchElementException:
    dataset_names.append("-")
    data_types.append("-")
    tasks.append("-")
    attribute_types.append("-")
    instances.append("-")
    attributes.append("-")
    years.append("-")
```
