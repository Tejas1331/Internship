Q1) Write a python program to display all the header tags from wikipedia.org.

```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.wikipedia.org.")

soup = BeautifulSoup(response.content)

header_tags = soup.find_all(['h1','h2','h3','h4','h5','h6'])

for tag in header_tags:
    print(tag.text.strip())
```

Q3) Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name
ii) Cuisine iii) Location iv) Ratings v) Image URL.

```python

## Restaurant names

import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.dineout.co.in/pune-restaurants/20instant-discount")

soup = BeautifulSoup(response.content)

restaurant_name = soup.find_all('a', class_="restnt-name ellipsis")

for tag in restaurant_name:
    print(tag.text.strip())
```

```python
## Restaurant Cuisines

import requests
from bs4 import BeautifulSoup
import re

response = requests.get("https://www.dineout.co.in/pune-restaurants/20instant-discount")

soup = BeautifulSoup(response.content)

restaurant_cuisine = soup.find_all('span', class_ ="double-line-ellipsis")

pattern = r'\| (\w+\s?\w+)'

for tag in restaurant_cuisine:
  x = re.search(pattern,tag.text.strip())
  print(x.group(1))
```

```python
## Location

import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.dineout.co.in/pune-restaurants/20instant-discount")

soup = BeautifulSoup(response.content)

restaurant_location = soup.find_all('div', class_='restnt-loc ellipsis')

for rest in restaurant_location:
  print(rest.text.strip())
```

```python
## Ratings

import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.dineout.co.in/pune-restaurants/20instant-discount")

soup = BeautifulSoup(response.content)

restaurant_location = soup.find_all('div', class_=['restnt-rating rating-5','restnt-rating rating-4', 'restnt-rating rating-3', 'restnt-rating rating-2'])

for rest in restaurant_location:
  print(rest.text.strip())
```

```python
# Image URL

import requests
from bs4 import BeautifulSoup

response = requests.get('https://www.dineout.co.in/pune-restaurants/20instant-discount')

soup = BeautifulSoup(response.content)

image_url = soup.find_all('img', class_="no-img")

pattern = r'(data-src=)(\S+)'

for image in image_url:
  img = re.search(pattern, str(image))
  print(img.group(2))

```

Q4) Write s python program to display list of respected former finance minister of India(i.e.
Name , Term of office) from https://presidentofindia.nic.in/former-presidents and make
data frame.

```python
# Names

import requests
from bs4 import BeautifulSoup

response = requests.get('https://presidentofindia.nic.in/former-presidents')

soup = BeautifulSoup(response.content)

name = soup.find_all('div', class_='desc-sec')

pattern = r'[^h3<>/]'

names =[]

pattern = r'<h3>(.*)</h3>'

for n in name:
  text = n.find('h3')
  x= re.search(pattern,str(text))
  names.append(x.group(1))
  print(x.group(1))
```

```python
# Tenure

import requests
from bs4 import BeautifulSoup

response = requests.get('https://presidentofindia.nic.in/former-presidents')

soup = BeautifulSoup(response.content)

name = soup.find_all('div', class_='desc-sec')

tenures =[]

pattern = r'<h5>(.*)</h5>'

for n in name:
  text = n.find('h5')
  x= re.search(pattern,str(text))
  tenures.append(x.group(1))
  print(x.group(1))
```

```python
import pandas as pd

df = pd.DataFrame({'Names': names,'Tenures': tenures})

df.head()
```
